```python
import math

board = [[' ' for _ in range(3)] for _ in range(3)]

def print_board(board):
    for row in board:
        print('|' + '|'.join(row) + '|')

def check_winner(b):
    for i in range(3):
        if b[i][0] == b[i][1] == b[i][2] != ' ': return b[i][0]
        if b[0][i] == b[1][i] == b[2][i] != ' ': return b[0][i]

    if b[0][0] == b[1][1] == b[2][2] != ' ': return b[0][0]
    if b[0][2] == b[1][1] == b[2][0] != ' ': return b[0][2]

    if all(cell != ' ' for row in b for cell in row):
        return 'Tie'

    return None

def minimax(b, depth, is_maximizing):
    result = check_winner(b)
    if result == 'O': return 10 - depth
    if result == 'X': return depth - 10
    if result == 'Tie': return 0

    if is_maximizing:
        best_score = -math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'O'
                    score = minimax(b, depth + 1, False)
                    b[i][j] = ' '
                    best_score = max(score, best_score)
        return best_score
    else:
        best_score = math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'X'
                    score = minimax(b, depth + 1, True)
                    b[i][j] = ' '
                    best_score = min(score, best_score)
        return best_score

def best_move(b):
    best_score = -math.inf
    move = (-1, -1)
    for i in range(3):
        for j in range(3):
            if b[i][j] == ' ':
                b[i][j] = 'O'
                score = minimax(b, 0, False)
                b[i][j] = ' '
                if score > best_score:
                    best_score = score
                    move = (i, j)
    return move

def play():
    print("Tic-Tac-Toe: Bạn (X) vs AI (O)")
    while True:
        print_board(board)
        winner = check_winner(board)
        if winner:
            if winner == 'Tie': print("Hòa!")
            else: print(f"Người chiến thắng: {winner}")
            break


        try:
            row, col = map(int, input("Nhập hàng và cột (0-2), vd '0 1': ").split())
            if board[row][col] != ' ':
                print("Ô đã có người đi, thử lại!")
                continue
            board[row][col] = 'X'
        except:
            print("Dữ liệu nhập không hợp lệ!")
            continue

        if not check_winner(board):
            ai_row, ai_col = best_move(board)
            board[ai_row][ai_col] = 'O'
            print(f"AI đã đi vào: {ai_row} {ai_col}")

if __name__ == "__main__":
    play()
```

```python
import math
import time

class TicTacToe:
    def __init__(self, n=5, win_length=4):
        self.n = n
        self.win_length = win_length
        self.board = [[' ' for _ in range(n)] for _ in range(n)]

    def print_board(self):
        print("  " + " ".join(str(i) for i in range(self.n)))
        for i, row in enumerate(self.board):
            print(f"{i} " + "|".join(row))

    def check_winner(self, b):

        for r in range(self.n):
            for c in range(self.n):
                if b[r][c] == ' ': continue

                for dr, dc in [(0, 1), (1, 0), (1, 1), (1, -1)]:
                    count = 0
                    for i in range(self.win_length):
                        nr, nc = r + dr*i, c + dc*i
                        if 0 <= nr < self.n and 0 <= nc < self.n and b[nr][nc] == b[r][c]:
                            count += 1
                        else: break
                    if count == self.win_length: return b[r][c]

        if all(cell != ' ' for row in b for cell in row): return 'Tie'
        return None

    def evaluate(self, b):
        """Hàm Heuristic: Tính điểm cho các chuỗi quân cờ"""

        score = 0

        return score

    def minimax(self, b, depth, alpha, beta, is_maximizing):
        winner = self.check_winner(b)
        if winner == 'O': return 1000 + depth
        if winner == 'X': return -1000 - depth
        if winner == 'Tie': return 0
        if depth == 0: return 0

        if is_maximizing:
            best_score = -math.inf
            for r in range(self.n):
                for c in range(self.n):
                    if b[r][c] == ' ':
                        b[r][c] = 'O'
                        score = self.minimax(b, depth - 1, alpha, beta, False)
                        b[r][c] = ' '
                        best_score = max(score, best_score)
                        alpha = max(alpha, score)
                        if beta <= alpha: break
            return best_score
        else:
            best_score = math.inf
            for r in range(self.n):
                for c in range(self.n):
                    if b[r][c] == ' ':
                        b[r][c] = 'X'
                        score = self.minimax(b, depth - 1, alpha, beta, True)
                        b[r][c] = ' '
                        best_score = min(score, best_score)
                        beta = min(beta, score)
                        if beta <= alpha: break
            return best_score

    def get_best_move(self):
        best_score = -math.inf
        move = None

        depth = 4 if self.n <= 5 else 3

        for r in range(self.n):
            for c in range(self.n):
                if self.board[r][c] == ' ':
                    self.board[r][c] = 'O'
                    score = self.minimax(self.board, depth, -math.inf, math.inf, False)
                    self.board[r][c] = ' '
                    if score > best_score:
                        best_score = score
                        move = (r, c)
        return move

n_size = int(input("Nhập kích thước bàn cờ N (vd 5): "))
win_size = int(input("Nhập số quân liên tiếp để thắng (vd 4): "))
game = TicTacToe(n=n_size, win_length=win_size)

while True:
    game.print_board()
    if game.check_winner(game.board): break


    r, c = map(int, input("Lượt bạn (hàng cột): ").split())
    if game.board[r][c] != ' ': continue
    game.board[r][c] = 'X'
    if not game.check_winner(game.board):
        print("AI đang tính toán...")
        start = time.time()
        m = game.get_best_move()
        if m:
            game.board[m[0]][m[1]] = 'O'
            print(f"AI đi ({m[0]}, {m[1]}) trong {time.time()-start:.2f}s")

winner = game.check_winner(game.board)
game.print_board()
print(f"Kết thúc: {winner}")
```

```python
import math
import time

class TicTacToeNxN:
    def __init__(self, n=5, win_len=4):
        self.n = n
        self.win_len = win_len
        self.board = [[' ' for _ in range(n)] for _ in range(n)]
        self.ai = 'O'
        self.human = 'X'

    def print_board(self):
        print("\n   " + "  ".join(f"{i:2d}" for i in range(self.n)))
        for i, row in enumerate(self.board):
            print(f"{i:2d} |" + "|".join(f" {cell} " for cell in row) + "|")

    def check_winner(self, b, r, c):
        """Kiểm tra thắng cục bộ xung quanh nước đi vừa đánh để tăng tốc"""
        player = b[r][c]
        directions = [(0, 1), (1, 0), (1, 1), (1, -1)]
        for dr, dc in directions:
            count = 1

            for direction in [1, -1]:
                step = 1
                while True:
                    nr, nc = r + dr * step * direction, c + dc * step * direction
                    if 0 <= nr < self.n and 0 <= nc < self.n and b[nr][nc] == player:
                        count += 1
                        step += 1
                    else: break
            if count >= self.win_len: return player
        if all(cell != ' ' for row in b for cell in row): return 'Tie'
        return None

    def get_nearby_moves(self):
        """Chỉ xét các ô trống nằm cạnh các quân đã đánh (Bán kính 1-2 ô)"""
        moves = set()
        for r in range(self.n):
            for c in range(self.n):
                if self.board[r][c] != ' ':
                    for dr in range(-1, 2):
                        for dc in range(-1, 2):
                            nr, nc = r + dr, c + dc
                            if 0 <= nr < self.n and 0 <= nc < self.n and self.board[nr][nc] == ' ':
                                moves.add((nr, nc))
        return list(moves) if moves else [(self.n//2, self.n//2)]

    def evaluate_board(self, b):
        """Hàm lượng giá heuristic sơ bộ (có thể mở rộng)"""
        return 0

    def minimax(self, b, last_r, last_c, depth, alpha, beta, is_max):
        winner = self.check_winner(b, last_r, last_c)
        if winner == self.ai: return 100 + depth
        if winner == self.human: return -100 - depth
        if winner == 'Tie' or depth == 0: return 0

        moves = self.get_nearby_moves()
        if is_max:
            best = -math.inf
            for r, c in moves:
                if b[r][c] == ' ':
                    b[r][c] = self.ai
                    score = self.minimax(b, r, c, depth - 1, alpha, beta, False)
                    b[r][c] = ' '
                    best = max(best, score)
                    alpha = max(alpha, best)
                    if beta <= alpha: break
            return best
        else:
            best = math.inf
            for r, c in moves:
                if b[r][c] == ' ':
                    b[r][c] = self.human
                    score = self.minimax(b, r, c, depth - 1, alpha, beta, True)
                    b[r][c] = ' '
                    best = min(best, score)
                    beta = min(beta, best)
                    if beta <= alpha: break
            return best

    def play(self):
        print(f"Bàn cờ {self.n}x{self.n}. Cần {self.win_len} quân để thắng.")
        while True:
            self.print_board()

            r, c = map(int, input("Nhập hàng và cột (vd: 2 3): ").split())
            if self.board[r][c] != ' ': continue
            self.board[r][c] = self.human
            if self.check_winner(self.board, r, c): break


            print("AI đang suy nghĩ...")
            start = time.time()
            best_score = -math.inf
            move = (0, 0)

            depth = 4 if self.n <= 5 else 2

            for nr, nc in self.get_nearby_moves():
                self.board[nr][nc] = self.ai
                score = self.minimax(self.board, nr, nc, depth, -math.inf, math.inf, False)
                self.board[nr][nc] = ' '
                if score > best_score:
                    best_score = score
                    move = (nr, nc)

            self.board[move[0]][move[1]] = self.ai
            print(f"AI đánh: {move} ({time.time()-start:.2f}s)")
            if self.check_winner(self.board, move[0], move[1]): break

        self.print_board()
        print("Trò chơi kết thúc!")


game = TicTacToeNxN(n=10, win_len=5)
game.play()
```

```python
import tkinter as tk
from tkinter import messagebox
import math

class TicTacToeGUI:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title("Tic-Tac-Toe AI (Alpha-Beta)")
        self.board = [' ' for _ in range(9)]
        self.human = 'X'
        self.ai = 'O'
        self.buttons = []
        self._create_widgets()

    def _create_widgets(self):
        # Tạo lưới 3x3 các nút bấm
        for i in range(9):
            btn = tk.Button(self.window, text='', font=('Arial', 24, 'bold'),
                           width=5, height=2,
                           command=lambda i=i: self.human_move(i))
            btn.grid(row=i//3, column=i%3)
            self.buttons.append(btn)

        # Nút Reset
        reset_btn = tk.Button(self.window, text="Chơi lại", command=self.reset_game)
        reset_btn.grid(row=3, column=0, columnspan=3, sticky="nsew")

    def human_move(self, i):
        if self.board[i] == ' ' and not self.check_winner(self.board):
            self.make_move(i, self.human)
            if not self.check_winner(self.board) and ' ' in self.board:
                self.window.after(500, self.ai_move) # AI suy nghĩ 0.5s cho tự nhiên

    def ai_move(self):
        move = self.get_best_move()
        self.make_move(move, self.ai)
        self.check_game_over()

    def make_move(self, i, player):
        self.board[i] = player
        color = "#2196F3" if player == 'X' else "#F44336"
        self.buttons[i].config(text=player, state=tk.DISABLED, disabledforeground=color)
        self.check_game_over()

    def check_game_over(self):
        winner = self.check_winner(self.board)
        if winner:
            if winner == 'Tie':
                messagebox.showinfo("Kết thúc", "Hòa rồi!")
            else:
                messagebox.showinfo("Kết thúc", f"{winner} đã chiến thắng!")
            self.reset_game()

    def check_winner(self, b):
        win_states = [(0,1,2), (3,4,5), (6,7,8), (0,3,6), (1,4,7), (2,5,8), (0,4,8), (2,4,6)]
        for s in win_states:
            if b[s[0]] == b[s[1]] == b[s[2]] != ' ': return b[s[0]]
        if ' ' not in b: return 'Tie'
        return None

    # --- Thuật toán Alpha-Beta Pruning ---
    def minimax(self, board, depth, alpha, beta, is_maximizing):
        res = self.check_winner(board)
        if res == self.ai: return 10 - depth
        if res == self.human: return depth - 10
        if res == 'Tie': return 0

        if is_maximizing:
            best = -math.inf
            for i in range(9):
                if board[i] == ' ':
                    board[i] = self.ai
                    score = self.minimax(board, depth + 1, alpha, beta, False)
                    board[i] = ' '
                    best = max(score, best)
                    alpha = max(alpha, best)
                    if beta <= alpha: break
            return best
        else:
            best = math.inf
            for i in range(9):
                if board[i] == ' ':
                    board[i] = self.human
                    score = self.minimax(board, depth + 1, alpha, beta, True)
                    board[i] = ' '
                    best = min(score, best)
                    beta = min(beta, best)
                    if beta <= alpha: break
            return best

    def get_best_move(self):
        best_score = -math.inf
        move = 0
        for i in range(9):
            if self.board[i] == ' ':
                self.board[i] = self.ai
                score = self.minimax(self.board, 0, -math.inf, math.inf, False)
                self.board[i] = ' '
                if score > best_score:
                    best_score = score
                    move = i
        return move

    def reset_game(self):
        self.board = [' ' for _ in range(9)]
        for btn in self.buttons:
            btn.config(text='', state=tk.NORMAL)

    def run(self):
        self.window.mainloop()

if __name__ == "__main__":
    game = TicTacToeGUI()
    game.run()
```
