```python
import heapq

def a_star(graph, start, goal, heuristic):
    open_set = []
    heapq.heappush(open_set, (0, start))

    came_from = {}
    g_score = {node: float('inf') for node in graph}
    g_score[start] = 0

    f_score = {node: float('inf') for node in graph}
    f_score[start] = heuristic(start, goal)

    while open_set:
        _, current = heapq.heappop(open_set)

        if current == goal:
            return reconstruct_path(came_from, current)

        for neighbor, cost in graph[current]:
            tentative_g = g_score[current] + cost

            if tentative_g < g_score[neighbor]:
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g
                f_score[neighbor] = tentative_g + heuristic(neighbor, goal)
                heapq.heappush(open_set, (f_score[neighbor], neighbor))

    return None
def reconstruct_path(came_from, current):
    path = [current]
    while current in came_from:
        current = came_from[current]
        path.append(current)
    path.reverse()
    return path

def heuristic(u, v):
    return 0

graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 2), ('E', 5)],
    'C': [('F', 3)],
    'D': [('G', 1)],
    'E': [('G', 2)],
    'F': [('G', 1)],
    'G': []
}

start = 'A'
goal = 'G'

path = a_star(graph, start, goal, heuristic)

print("Đường đi:", path)
```

```python
import heapq


def a_star_tsp(graph, start):
    n = len(graph)
    all_nodes = set(graph.keys())


    open_set = []
    heapq.heappush(open_set, (0, 0, start, frozenset([start]), [start]))

    while open_set:
        f, g, current, visited, path = heapq.heappop(open_set)

        if len(visited) == n:
            if start in graph[current]:
                total_cost = g + graph[current][start]
                return path + [start], total_cost
        for neighbor in graph[current]:
            if neighbor not in visited:
                new_visited = visited | {neighbor}
                new_g = g + graph[current][neighbor]
                h = heuristic(graph, neighbor, new_visited, start)
                heapq.heappush(open_set,
                               (new_g + h, new_g, neighbor, new_visited, path + [neighbor]))

    return None, float('inf')

def heuristic(graph, current, visited, start):
    unvisited = set(graph.keys()) - visited
    if not unvisited:
        return graph[current].get(start, 0)

    min_edge = float('inf')
    for u in graph:
        for v in graph[u]:
            if u != v:
                min_edge = min(min_edge, graph[u][v])

    return min_edge * len(unvisited)

graph = {
    'A': {'B': 10, 'C': 15, 'D': 20},
    'B': {'A': 10, 'C': 35, 'D': 25},
    'C': {'A': 15, 'B': 35, 'D': 30},
    'D': {'A': 20, 'B': 25, 'C': 30}
}

start = 'A'

path, cost = a_star_tsp(graph, start)

print("Đường đi:", path)
print("Chi phí:", cost)
```
