```python
import random

def tao_ma_tran(m, n):
    return [[random.randint(1, 20) for _ in range(n)] for _ in range(m)]

def in_ma_tran(a):
    for row in a:
        print(*row)

def xuat_dong(a, k):
    print("Dong", k, ":", a[k])

def xuat_cot(a, k):
    print("Cot", k, ":",end=" ")
    for row in a:
        print(row[k], end=" ")
    print()

def dong_max_45(a):
    max_sum = -1
    index = -1
    for i in range(len(a)):
        s = sum(a[i])
        if s <= 45 and s > max_sum:
            max_sum = s
            index = i
    return index, max_sum

def cot_tich_min(a):
    n = len(a[0])
    min_tich = float('inf')
    index = -1
    for j in range(n):
        tich = 1
        for i in range(len(a)):
            tich *= a[i][j]
        if tich < min_tich:
            min_tich = tich
            index = j
    return index, min_tich

def dong_chan_cot_le(a):
    print("Phan tu dong chan, cot le:")
    for i in range(len(a)):
        for j in range(len(a[0])):
            if i % 2 == 0 and j % 2 == 1:
                print(a[i][j], end=" ")
    print()

def tbc_chan_dong_le(a):
    tong = dem = 0
    for i in range(len(a)):
        if i % 2 == 1:
            for x in a[i]:
                if x % 2 == 0:
                    tong += x
                    dem += 1
    return tong / dem if dem != 0 else 0

def tbc_bien(a):
    tong = dem = 0
    m, n = len(a), len(a[0])
    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                tong += a[i][j]
                dem += 1
    return tong / dem

def tbc_tich_trong(a):
    tich = 1
    dem = 0
    m, n = len(a), len(a[0])
    for i in range(1, m-1):
        for j in range(1, n-1):
            tich *= a[i][j]
            dem += 1
    return tich / dem if dem != 0 else 0

m = int(input("Nhap so dong m: "))
n = int(input("Nhap so dong n: "))

a = tao_ma_tran(m, n)

print("\nMa tran:")
in_ma_tran(a)

k = int(input("\nNhap k (dong/cot): "))

xuat_dong(a, k)
xuat_cot(a, k)
idx, s = dong_max_45(a)
print("Dong co tong lon nhat <=45:", idx, "Tong:", s)
idx, t = cot_tich_min(a)
print("Cot co tich nho nhat:", idx, "Tich:", t)
dong_chan_cot_le(a)

print("TBC chan dong le:", tbc_chan_dong_le(a))

print("TBC phan tu bien:", tbc_bien(a))

print("TBC tich phan tu trong:", tbc_tich_trong(a))
```

```python
from datetime import datetime

class SinhVien:
    def __init__(self, ma, ten, nam_sinh, diem_tb):
        self.ma = ma
        self.ten = ten
        self.nam_sinh = nam_sinh
        self.diem_tb = diem_tb

    def __str__(self):
        return f"{self.ma} - {self.ten} - {self.nam_sinh} - {self.diem_tb}"


def nhap_ds(n):
    ds = []
    for i in range(n):
        print(f"\nNhap sinh vien {i+1}:")
        ma = input("Ma SV: ")
        ten = input("Ten: ")
        nam_sinh = int(input("Nam sinh: "))
        diem_tb = float(input("Diem TB: "))
        ds.append(SinhVien(ma, ten, nam_sinh, diem_tb))
    return ds

def dem_len_lop(ds):
    return sum(1 for sv in ds if sv.diem_tb >= 5)

def sv_20_tuoi(ds):
    nam_hien_tai = datetime.now().year
    print("\nSinh vien du 20 tuoi:")
    for sv in ds:
        if nam_hien_tai - sv.nam_sinh >= 20:
            print(sv)

def dem_he_dh(ds):
    return sum(1 for sv in ds if len(sv.ma) >= 4 and sv.ma[2:4] == "DH")

def dem_ten_lan(ds):
    return sum(1 for sv in ds if sv.ten.split()[-1].lower() == "lan")

def dem_ho_phan(ds):
    return sum(1 for sv in ds if sv.ten.split()[0].lower() == "phan")

n = int(input("Nhap so sinh vien: "))
ds = nhap_ds(n)

print("\nDanh sach sinh vien:")
for sv in ds:
    print(sv)

print("\nSo SV du dieu kien len lop:", dem_len_lop(ds))

sv_20_tuoi(ds)

print("So SV he dai hoc:", dem_he_dh(ds))

# f
print("So SV ten Lan:", dem_ten_lan(ds))

# g
print("So SV ho Phan:", dem_ho_phan(ds))
```

```python
def tron_mang(a, b):
    c = []
    i = 0

    while i < len(a) and i < len(b):
        c.append(min(a[i], b[i]))
        i += 1

    while i < len(a):
        c.append(a[i])
        i += 1

    while i < len(b):
        c.append(b[i])
        i += 1

    return c

a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

ket_qua = tron_mang(a, b)
print("Mảng kết quả:", ket_qua)
```

```python
import random

def tao_ma_tran(m, n):
    a = []
    for i in range(m):
        row = []
        for j in range(n):
            row.append(random.randint(0, 100))
        a.append(row)
    return a

def xuat_dong(a, k):
    print("Dong", k, ":", a[k])


def xuat_cot(a, k):
    print("Cot", k, ":")
    for i in range(len(a)):
        print(a[i][k], end=" ")
    print()

def dong_tong_lon_hon_45(a):
    for i in range(len(a)):
        if sum(a[i]) > 45:
            print("Dong", i, "co tong >", 45, ":", a[i])


def cot_tich_nho_nhat(a):
    m = len(a)
    n = len(a[0])
    min_tich = float('inf')
    cot_min = -1

    for j in range(n):
        tich = 1
        for i in range(m):
            tich *= a[i][j]
        if tich < min_tich:
            min_tich = tich
            cot_min = j

    print("Cot co tich nho nhat la:", cot_min)

def dong_chan_cot_le(a):
    print("Phan tu dong chan, cot le:")
    for i in range(len(a)):
        for j in range(len(a[0])):
            if i % 2 == 0 and j % 2 == 1:
                print(a[i][j], end=" ")
    print()

def tbc_chan_dong_le(a):
    tong = 0
    dem = 0
    for i in range(len(a)):
        if i % 2 == 1:
            for j in range(len(a[0])):
                if a[i][j] % 2 == 0:
                    tong += a[i][j]
                    dem += 1
    if dem > 0:
        print("TBC =", tong / dem)
    else:
        print("Khong co phan tu hop le")

def tbc_bien(a):
    m = len(a)
    n = len(a[0])
    tong = 0
    dem = 0

    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                tong += a[i][j]
                dem += 1

    print("TBC bien =", tong / dem)

m, n = 4, 5
a = tao_ma_tran(m, n)

print("Ma tran:")
for row in a:
    print(row)

xuat_dong(a, 1)
xuat_cot(a, 2)
dong_tong_lon_hon_45(a)
cot_tich_nho_nhat(a)
dong_chan_cot_le(a)
tbc_chan_dong_le(a)
tbc_bien(a)
```
